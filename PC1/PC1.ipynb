##Funciones

```python
def MostrarValores(n, m, jobs):
  print("La cantidad de tareas es: ", n)
  print("La cantidad de máquinas es: ", m)
  print("Tareas/Jobs por máquina: ")
  print(jobs)
```

```python
def LeerArchivo(nombre):
  
  with open(nombre) as f:
    n, m = [int(x) for x in f.readline().split()]
    jobs = np.zeros((n, m, 2), dtype=int)
    i = 0
    for line in f:
        jobs[i] = np.array([int(x) for x in line.split()]).reshape(m, 2)
        i += 1

  return n, m, jobs
```

```python
def MakeSpan(n, m, rep):
  mstart = np.zeros((m), dtype=int) #Vector que contiene el inicio de cada una de las máquinas
  jend = np.zeros((m), dtype=int) #Vector que contiene el tiempo final en que se terminaron las tareas
  idxs = np.zeros((n), dtype=int) 
  makespan = 0 #Variable que contiene el tiempo total 

  for e in rep:
    i = idxs[e]
    idxs[e] += 1
    midx = jobs[e, i, 0]
    time = jobs[e, i, 1]
    #print(f"Job {e} task {i} to machine {midx} starting at {mstart[midx]} duration {time}")
    mstart[midx] = max(mstart[midx], jend[e]) + time
    jend[e] = mstart[midx]
    if mstart[midx] > makespan:
        makespan = mstart[midx]

  return makespan
```

```python
def Horario(n, m, rep, makespan):
  mstart = np.zeros((m), dtype=int)
  jend = np.zeros((m), dtype=int)
  idxs = np.zeros((n), dtype=int)
  schedule = np.zeros((m*10, makespan))

  for e in rep:
    i = idxs[e]
    idxs[e] += 1
    midx = jobs[e, i, 0]
    time = jobs[e, i, 1]
    mstart[midx] = max(mstart[midx], jend[e])
    schedule[midx*10:(midx+1)*10, mstart[midx]:mstart[midx] + time] = e+1
    mstart[midx] += time
    jend[e] = mstart[midx]
  
  return schedule
```

```python
#Función que corre todo: obtiene el makespan y te muestra el horario
def Final(Archivo):
  n, m, jobs = LeerArchivo(Archivo) #Leer el archivo y obtener los valores de: n, m, jobs
  MostrarValores(n, m, jobs) #Mostrar valores

  rep = [] #Vector de rep
  for i in range(n):
    rep.extend(np.full(m,i)) #añadir i veces el valor de la máquina m
  print(rep)

  np.random.shuffle(rep) #Volver rep random
  print("rep: ", rep) #Mostrar rep

  makespan = MakeSpan(n, m, rep) #Obtener el Makespan
  print("El tiempo total para que todas las máquinas realicen todas las tareas es: ", makespan)

  schedule = Horario(n, m, rep, makespan) #Obtener el horario
  plt.imshow(schedule) #Mostrar el horario
```

```python
#Función que sólo obtiene el makespan que se usará para correrlo 3 minutos
def Final3Min(Archivo):
  n, m, jobs = LeerArchivo(Archivo) #Leer el archivo y obtener los valores de: n, m, jobs
  MostrarValores(n, m, jobs) #Mostrar valores (probar que funcione, se comenta al final)

  rep = [] #Vector de rep
  for i in range(n):
    rep.extend(np.full(m,i)) #añadir i veces el valor de la máquina m
  print(rep)
  random.shuffle(rep) #Volver rep random
  print("rep: ", rep) #Mostrar rep

  makespan = MakeSpan(n, m, rep) #Obtener el Makespan
  print("El tiempo total para que todas las máquinas realicen todas las tareas es: ", makespan) #prueba, se comenta al final

  return makespan
```

```python
def Minutos3(Archivo):
  seg_totales = datetime.datetime.now().second + 180 #Para obtener 3 minutos (60 * 3 = 180 segundos = 3 minutos)
  RandomSamplingRespuestas = [] #Vector que guardará las respuestas de RandomSampling para mostrarlas en el gráfico
  tiempo = [] #Vector del tiempo en el eje y
  segundos = 0

  while datetime.datetime.now().second < seg_totales:
    makespan = Final3Min(Archivo) #Ejecutar el algoritmo
    RandomSamplingRespuestas.append(makespan) #Agregar el makespan obtenido
    tiempo.append(segundos)
    segundos += 1
    #time.sleep(0.3)
  
  plt.plot(tiempo, RandomSamplingRespuestas)
  plt.show()

```

###Implementar el algoritmo Simulated Annealing:

**Sobre el Algoritmo:**
* Es un algoritmo de búsqueda metaheurística para problemas de optimización global.

* **Objetivo:** Encontrar una buena aproximación al valor óptimo de una función en un espacio de búsqueda grande. 
  - El "óptimo global" es la solución al problema y no hay un mejor valor que este.
  - Si el problema es de minimización, el óptimo global es el valor más pequeño posible de todos los valores del espacio de búsqueda.
  - Si el problema es de optimización, el óptimo global es el valor más grande de todos los valores del espacio de búsqueda.

* **Origen del Algoritmo:** el nombre del algoritmo viene del proceso de recocido del acero y cerámica, técnica que consiste en calentar y luego enfriar lentamente el material para ver sus propiedades físicas. Por el calor, los átomos se desplazan de sus posiciones iniciales (mínimo local de energía). Con el enfriamiento se obtienen las probabilidades de recristalizar, el enfriamiento lento permite obtener mayores probabilidades de recristalizar en configuraciones con menor energía que la inicial (mínimo global). Este método fue descrito por  Scott Kirkpatrick, C. Daniel Gelatt y Mario P. Vecchi en 1983 y por Vlado Černý en 1985. No obstante, el método de Simulated Annealing es una adaptación de un método de Montecarlo utilizado para genera muestras de estados de un sistema termodinámico.

* **Aplicaciones:** Este algoritmo tiene múltiples aplicaciones pero es ideal para los siguientes problemas:
  - Campo de la optimización de estructuras de hormigón
  - Resolver instancias del problema ([[Job Shop Scheduling]])
